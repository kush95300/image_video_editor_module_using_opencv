```python
import cv2
import os
import numpy as np
import random
from tkinter import * 
from tkinter.ttk import *
```

```python
def screen_fit():
    
    root=Tk()
    h=root.winfo_screenheight()
    w=root.winfo_screenwidth()
    return (w,h)
```

```python
def add_text_to_image(image ,message="your msg",font_type=0,font_size=1,color="green",line_type=4,message_postion=1,thickness=2,move=10):
    if(line_type==0):
        line_type=cv2.FILLED
    elif line_type==1:
        line_type=cv2.LINE_4
    elif line_type==2:
        line_type=cv2.LINE_8
    else:
        line_type=cv2.LINE_AA
    
    if color=="random":
        color=(random.randint(0,255),random.randint(0,255),random.randint(0,255))
    elif color=='black' or color==0:
        color=(0,0,0)
    elif color=="white" or color==1:
        color=(255,255,255)
    elif color=="pink" or color==2:
        color=(255,0,255)
    elif color=="yellow" or color==3:
        color=(0,255,255)
    elif color=="cyan" or color==4:
        color=(255,255,0)
    elif color=="blue" or color==5:
        color=(255,0,0)
    elif color=="green" or color==6:
        color=(0,255,0)
    else:
        color=(0,0,255)
        
    shape=image.shape
    
        
    if  message_postion % 3 == 0:
        message_postion=(move,30)
    elif message_postion % 3 == 1:
        message_postion=(move,int((shape[1]/2)))
    else :
        message_postion=(move,shape[1])

        
    return cv2.putText(image,message,message_postion, font_type, font_size,color,thickness,line_type)

```

```python
def mirror(camara_no,size_type=1, text_move=True, text_moving_speed=10):
        #camara on
        if(text_move):
            i=10
        else:
            i=15
        lappy=cv2.VideoCapture(camara_no)
        while True:
            status1,pic1=lappy.read()
            if(status1==False):
                print("Camara not Connected..... Something Error...")
            else:
                if size_type==0:
                    pic=pic1
                elif size_type==1:
                    pic=cv2.resize(pic1,screen_fit())
                elif size_type==2:
                    pic=cv2.resize(pic1,(1500,900))
                elif size_type==3:
                    pic=cv2.resize(pic1,(1200,850))
                elif size_type==4:
                    pic=cv2.resize(pic1,(750,600))
                elif size_type==5:
                    pic=cv2.resize(pic1,(700,500))
                elif size_type==6:
                    pic=cv2.resize(pic1,(600,600))
                elif size_type==7:
                    pic=cv2.resize(pic1,(500,500))
                elif size_type==8:
                    pic=cv2.resize(pic1,(500,400))
                elif size_type==9:
                    pic=cv2.resize(pic1,(400,400))
                elif size_type==10:
                    pic=cv2.resize(pic1,(400,280))
                elif size_type==11:
                    pic=cv2.resize(pic1,(350,350))
                elif size_type==12:
                    pic=cv2.resize(pic1,(350,300))
                elif size_type==13:
                    pic=cv2.resize(pic1,(350,250))
                
                if(i>pic.shape[1]-100):
                    i=0
                
                
                pic=add_text_to_image(image=pic,message="Press 'esc' key to stop",font_size=1,color='random', move=i)
                
                cv2.imshow("Friend_Editor-Mirror",pic)
                i+=text_moving_speed
                if cv2.waitKey(1) == 27:
                    break
        cv2.destroyAllWindows()
        lappy.release()    
```
